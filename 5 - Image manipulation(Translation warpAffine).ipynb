# Translation
This is an affine transform that simpy shifts the position of an image.
We use cv2.warpAffine to implement these transformations

# warpAffine

### cv2.warpAffine(image, Translation matrix, (width, height))

#### Translation matrix

```python
Image('tutorial images/translation matrix.png')
```

```python
import cv2, numpy as np
image = cv2.imread('my.JPG')
#store height and width of the image
height, width = image.shape[:2]
quarter_height, quarter_width = height / 4, width / 4

#      [1 0 Tx]
# T = [0 1 Ty]

#T is our transition matrix
T = np.float32([[1, 0, -quarter_width], [0, 1, -quarter_height]])

#We use warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (width, height))
cv2.imshow('Translation', img_translation)
#cv2.imwrite('Translation.png', img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```
